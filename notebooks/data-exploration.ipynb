## Multivariate Analysis

### 1) Fuel Consumption and CO2 Emission

```python
plt.figure(figsize=(10, 10))

plt.subplot(2,2,1)
sns.scatterplot(x=df['Fuel Consumption City (L/100 km)'], y=df['CO2 Emissions(g/km)'])

plt.subplot(2,2,2)
sns.scatterplot(x=df['Fuel Consumption Hwy (L/100 km)'], y=df['CO2 Emissions(g/km)'])

plt.subplot(2,2,3)
sns.scatterplot(x=df['Fuel Consumption Comb (L/100 km)'], y=df['CO2 Emissions(g/km)'])

plt.subplot(2,2,4)
sns.scatterplot(x=df['Fuel Consumption Comb (mpg)'], y=df['CO2 Emissions(g/km)'])
```

```python
cols = [
    'Fuel Consumption City (L/100 km)',
    'Fuel Consumption Hwy (L/100 km)',
    'Fuel Consumption Comb (L/100 km)',
    'Fuel Consumption Comb (mpg)'
]

df[cols].corr()
```

- sort of linear relation between Fuel Consumption and CO2 Emissions
- upper side vehicles are more agressive to the emission 
- these 3 are highly positively correlated
- and last one is highly negetively correlated 



### 2) Fuel Type and CO2 Emission

```python
df['CO2 Emissions(g/km)'].groupby(df['Fuel Type']).mean()
```

- Ethanol based vehicles are more agressive towards CO2 Emission
